Making My Third Project Named 'Number Guessing Game'.

```python
import random

best_score = 0
while True:
    print("Select difficulty level:")
    print("1. Easy(1-50)")
    print("2. Medium(50-100)")
    print("3. Hard(1-200)")

    level = input("Enter your level(1/2/3): ")

    if level == "1":
        max_number = 50
        max_attempts = 10
    elif level == "2":
        max_number = 100 
        max_attempts = 7
    elif level == "3":
        max_number = 200
        max_attempts = 5
    else:
        print("Invalid choice, defaulting to Medium")
        max_attempts = 7
    
    computer = random.randint(1,100)
    user = 0
    no_of_attempts = 0

    while user != computer:
        if no_of_attempts >= max_attempts  :
            print(f"you exceeded the limit of {max_attempts}")
            print("Game Over")
            break
        user = int(input("Guess a number between 1 and 100: "))
    
        if user < 1 or user > 100:
            print("Please enter between 1 and 100")
            continue
            
        no_of_attempts += 1
 
        if user < computer:
            print("Too Low!!")
        elif user > computer:
            print("Too High!!")
    
        else:
            print("Correct")
            print(f"You guessed it in  {no_of_attempts} attempts!")
            break

        difference = abs(user - computer)

        if difference <= 5:
            print("Very close!")
        elif difference <= 10:
            print("close")
        else:
            print("Far away")

        score = max_attempts - no_of_attempts + 1
        print("your score is:",score)

        if score > best_score:
            best_score = score
        print("Your best_score:" , best_score)
        
        

    replay = input("Do you want to play again? (yes or no): ").lower()
    if replay == "no":
        print("Thanks for playing!")
        break
```
