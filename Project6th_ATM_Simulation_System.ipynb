```python
#3  ATM system
# Data: balance,pin
#action: check_balance(pin) , deposit(pin,amount) , withdraw(pin,amount)
#rules: if wrong pin -> access denied
#       cannot withdraw more than balance
# ATM simulation system
import os
from datetime import datetime
class ATM_machine:
    def __init__(self,balance,pin):
        self.__balance = balance  # private
        self.__pin = pin
        self.__attempts = 0   
        self.history = []

    def save_data(self):
        with open("/home/661d5863-d774-4ee3-8570-4334183d995c/atm.txt", "a") as file:
            file.write(str(self.__balance) + "\n")
            file.write(str(self.__pin) + "\n")

            for h in self.history:
                file.write(h + "\n")
        print("File saved successfully!")
        

    def change_pin(self,old_pin,new_pin):
        if self.__check_pin(old_pin):
            self.__pin = new_pin
            print("Pin Changed Successfully")
        else:
            print("Wrong Old Pin")

    def show_history(self):
        if len(self.history) == 0:
            print("No transaction yet")
        else:
            for h in self.history:
                print(h)

    def check_pin(self, pin):
        if self.__attempts >= 3:
            print("Account Blocked")
            return False

        if self.__pin == pin:
            self.__attempts = 0
            return True
        else:
            self.__attempts += 1
            print("Incorrect pin")
            return False

    def check_balance(self,pin):
        if self.check_pin(pin):
            time = datetime.now().strftime("%d-%m-%Y %H:%M")
            print(f"Your balance is: {self.__balance}")
            self.history.append(f"{time} - Balance checked: {self.__balance}")
        

    def deposit(self,pin,amount): 
        if self.check_pin(pin):
            self.__balance += amount
            time = datetime.now().strftime("%d-%m-%Y %H:%M")
            print(f"{amount} deposited at {time}. New balance: {self.__balance}")
            self.history.append(f"{time} - Deposited {amount} ")
        
    def withdraw(self,pin,amount):
        if self.check_pin(pin):
            if amount <= self.__balance:
                self.__balance -= amount
                time = datetime.now().strftime("%d-%m-%Y %H:%M")
                print(f"{amount} withdrawn at {time}. Remaining balance: {self.__balance}")
                self.history.append(f"{time} - withdrew {amount} ")
            else:
                print("insufficient balance!!")


A1 = ATM_machine(5000 , 1234)

A1.deposit(1234 , 3000)
A1.withdraw(1234 , 1000)
A1.check_balance(1234)


pin = int(input("Enter PIN: "))

if pin == 1234:
    print("Login Successfully")
    
    while True:
        print("\n1. Check Balance")
        print("2.Deposit")
        print("3.Withdraw")
        print("4.Exit")
        print("5.Show History")
        print("6.Change PIN")

        choice = int(input("Enter your choice: "))

        if choice == 1:
            pin = int(input("Enter PIN: "))
            A1.check_balance(pin)

        elif choice == 2:
            pin = int(input("Enter PIN: "))
            amount = int(input("Enter the amount: "))
            A1.deposit(pin , amount)
            A1.save_data()
        elif choice == 3:
            pin = int(input("Enter PIN: "))
            amount = int(input("Enter the amount: "))
            A1.withdraw(pin , amount)
            A1.save_data()
        elif choice == 4:
            A1.save_data()
            print("Data saved. Bye!")
            print("Thankyou for exiting...")
            break
        elif choice == 5:
            A1.show_history()
        elif choice == 6:
            old = int(input("Enter old pin: "))
            new = int(input("Enter new pin: "))
            A1.change_pin(old,new)
            A1.save_data()
        else:
            print("Invalid choice")
    

print(os.getcwd())  
print(A1.history)
    
```
