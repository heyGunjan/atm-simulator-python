Making My First Project Named 'Calculator'.

```python


def add(a,b):
    return a+b
def subtract(a,b):
    return a-b
def multiply(a,b):
    return a*b
def divide(a,b):
    if b != 0:
       return a/b
    else:
        return "Error:Division By Zero"
    


def calculator():
    user1 = float(input("Enter the first number: "))
    user2= float(input("Enter the second number: "))
    operator = input("Enter operator(+,-,*,/): ")

    if operator=="+":
        result = add(user1,user2)
        print("Result:", result)
    elif operator=="-":
        result = subtract(user1,user2)
        print("Result:", result)
    elif operator=="*":
        result = multiply(user1,user2)
        print("Result:", result)
    elif operator=="/":
        result = divide(user1,user2)
        print("Result:", result)
    else:
        print("Invalid Operator") 

while True:
    calculator()

    choice = input("Do you want to continue? (yes/no): ")

    if choice.lower != "yes":
        print("Calculator Closed.")
        break

calculator()


    
    
```
